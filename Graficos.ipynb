```python
import matplotlib.pyplot as plt
import numpy as np

x1 = ['Gelo', 'Neve', 'Rocha', 'Crostas biológicas', 'Bofedales']
y1 = [47410.00, 89383.00, 114519.00, 41321.00, 9876.00]
y2 = [38191.00, 52008.00, 173168.00, 32256.00, 6888.00]

total_y1 = sum(y1)
total_y2 = sum(y2)

y1_percent = [(valor / total_y1) * 100 for valor in y1]
y2_percent = [(valor / total_y2) * 100 for valor in y2]

plt.figure(figsize=(12, 6))


bar_width = 0.3
space = 0.1
index = np.arange(len(x1))


plt.bar(index, y1, bar_width, label="Junho", color='#0072B8')
plt.bar(index + bar_width + space, y2, bar_width, label="Outubro", color='#990000')

plt.ylabel('Área (km²)')

for i in range(len(y1)):

    if i == 2:
        plt.text(i, y1[i] + 2500, f'{y1_percent[i]:.1f}%', ha='center', va='bottom')
        plt.text(i + bar_width + space, y2[i] + 2500, f'{y2_percent[i]:.1f}%', ha='center', va='bottom')
    else:
        plt.text(i, y1[i] + 5000, f'{y1_percent[i]:.1f}%', ha='center', va='bottom')
        plt.text(i + bar_width + space, y2[i] + 5000, f'{y2_percent[i]:.1f}%', ha='center', va='bottom')
plt.xticks(index + (bar_width + space) / 2, x1)

plt.grid(axis='y', linestyle='-', alpha=0.7)

plt.tight_layout()
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=4)
plt.savefig('grafico_area_km2.png', format='png', dpi=300, bbox_inches='tight')

plt.show()
```
